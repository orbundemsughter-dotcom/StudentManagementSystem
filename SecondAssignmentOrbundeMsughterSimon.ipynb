# TASK A: STUDENT RESULT MANAGEMENT SYSTEM

```python
#Function To Calculate Grade

students = {}

def calculate_grade(score):
    if 70 <= score <= 100:
        return "A"
    elif 60 <= score <= 69:
        return "B"
    elif 50 <= score <= 59:
        return "C"
    elif 45 <= score <= 49:
        return "D"
    else:
        return "F"
```

```python
#To Add Student

def add_student():
    name = input("Enter Student name: ")
    try:
        score = float(input("Enter Student Score(0-100): "))
        if score < 0 or score > 100:
            print("Score must be between 0 and 100.")
            return
    
        students[name] = score
        print("Student added succesfully.")
    except ValueError:
        print("Invalid Score entered.")
```

```python
#View Students

def view_students():
    if not students:
        print("No student records available.")
        return
    print("\nStudent Records")
    print("-" * 40)

    for name, score in students.item():
        grade = calculate_grade(score)
        print(f"Name: {name}, score: {score}, Grade: {grade}")
```

```python
#Update Score

def update_score():
    
    name = input("Enter Student name to update: ")
    if name not in students:
        print("Student not found.")
        return
    try:
        score = float(input("Enter new score: "))
        if score < 0 or score > 100:
            print("Score must be between 0 and 100.")
            return
        students[name] = score
        print("Score updated successfully.")
    except ValueError:
        print("Invalid score entered.")
```

```python
#Search Student

def search_student():
    name = input("Enter student name to search: ")
    if name in students:
        score = students[name]
        grade = calculate_grade(score)
        print("\nStudent Found")
        print(f"Name: {name}")
        print(f"Score: {score}")
        print(f"Grade: {grade}")
    else:
        print("Student not found.")
```

```python
#Class Statistics

def class_statistics():
    if not students:
        print("No student records available.")
        return
        scores = list(students.value())
        print("\nClass Statistics")
        print("-" * 30)
        print("Highest Score: ", max(scores))
        print("Lowest Score: ", min(scores))
        print("Average Score: ", sum(scores)/len(scores))
        print("Total Students: ", len(scores))
```

```python
# Main Menu

while True:
    print("\n=====STUDENT RESULT MANAGEMENT SYSTEM=====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Update Score")
    print("4. Search Student")
    print("5. Class Statistics")
    print("6. Exit")

    choice = input("Enter choice: ")
    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        update_score()
    elif choice == "4":
        search_student()
    elif choice == "5":
        class_statistics()
    elif choice == "6":
        print("Exiting program...")
        break
    else:
        print("Invalid choice")
```

# TASK B: INVENTORY MANAGEMENT SYSTEM

```python
#Add Product

inventory = {}

def add_product():
    product = input("Enter product name: ")
    try:
        quantity = int(input("Enter quantity: "))
        if quantity < 0:
            print("Quantity cannot be negative.")
            return

        inventory[product] = quantity
        print("Product added successfully.")

    except ValueError:
        print("Invalid quantity entered.")
                       
        
                       
```

```python
#View Inventory

def view_inventory():
    if not inventory:
        print("Inventory is empty.")
        return

    print("\nInventory Records")
    print("-" * 30)

    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
```

```python
#Update Quantity

def update_quantity():
    product = input("Enter product name to update: ")
    if product not in inventory:
        print("Product not found.")
        return
    try:
        quantity = int(input("Enter new quantity: "))
        if quantity < 0:
            print("Quantity cannot be negative.")
            return
        Inventory[product] = quantity
        print("Quantity updated successfully.")
    except ValueError:
        print("Invalid quantity entered.")
```

```python
#Search Product

def search_product():
    product = input("Enter product name to search. ")
    if product in inventory:
        print(f"{product}: {inventory[product]}")
    else:
        print("Product not found.")
                    
```

```python
#Remove Product

def remove_product():
    product = input("Enter product to remove: ")
    if product in inventory:
        del inventory[product]
        print("Product removed successfully.")
    else:
        print("Product not found.")
```

```python
#Low Stock Report

def low_stock_report():
    print("\nLow Stock Products(Quantity < 5)")
    print("-" * 35)

    found = False
    for product, quantity in inventory.items():
        if quantity < 5:
            print(f"{product}: {quantity}")
            found = True
    if not found:
        print("No low stock items.")
```

```python
#Inventory Statistics

def inventory_statistics():
    if not inventory:
        print("Inventory is empty.")
        return
    quantities = list(inventory.values())
    highest_product = max(inventory, key = inventory.get)
    lowest_product = min(inventory, key = inventory.get)

    print("\nInventory Statistics")
    print("-" * 30)
    print("Total products:", len(inventory))
    print(f"Highest Quantity product: {highest_product} ({inventory[highest_product]})")
    print(f"Lowest Quantity product: {lowest_product} ({inventory[lowest_product]})")
    print("Total Stock available: ", sum(quantities))
```

```python
#Main Menu

while True:
    print("\n=====INVENTORY MANAGEMENT SYSTEM=====")
    print("1. Add Product")
    print("2. View Inventory")
    print("3. Update Quantity")
    print("4. Search Product")
    print("5. Remove Product")
    print("6. Low Stock Report")
    print("7. Inventory Statistics")
    print("8. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_product()
    elif choice == "2":
        view_inventory()
    elif choice == "3":
        update_quantity()
    elif choice == "4":
        search_product()
    elif choice == "5":
        remove_product()
    elif choice == "6":
        low_stock_report()
    elif choice == "7":
        inventory_statistics()
    elif choice == "8":
        print("Exiting program...")
        break
    else:
        print("Invalid choice.")
```
